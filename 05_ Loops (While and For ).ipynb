```python
n = int(input( 'Enter the value of n: ' ) )
print( 'Calculating factorial of', n, '...' )
result =1
i = 1
while i <= n:
  result = result * i
  print( 'i:', i, 'result:', result )
  i += 1
print("factorial of", n, "is", result)
```

```python
total = 0
counter =0
user_inp = int(input())
print(user_inp)

while user_inp > 0:   # Get values until 0 (or Less)

      total = total + user_inp
      counter = counter + 1
      user_inp = int(input())
      print(user_inp)

print(f'Average: {total / counter:.2f}\n')
```
